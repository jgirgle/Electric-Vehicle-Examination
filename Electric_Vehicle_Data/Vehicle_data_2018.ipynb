```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import scipy.stats as st
import numpy as np
import scipy.stats as linregress
```

```python
csv_path = "Resources/raw_data/vehicle_data_2018.csv"
vehicle_df_2018 = pd.read_csv(csv_path)
vehicle_df_2018
```

```python
vehicle_df_2018.drop(vehicle_df_2018.index[vehicle_df_2018['Model Year'] == '<2006'], inplace = True)
vehicle_df_2018.drop(vehicle_df_2018.index[vehicle_df_2018['Model Year'] == '2006'], inplace = True)
vehicle_df_2018.drop(vehicle_df_2018.index[vehicle_df_2018['Model Year'] == '2007'], inplace = True)
vehicle_df_2018.drop(vehicle_df_2018.index[vehicle_df_2018['Model Year'] == '2008'], inplace = True)
vehicle_df_2018.drop(vehicle_df_2018.index[vehicle_df_2018['Model Year'] == '2019'], inplace = True)

vehicle_df_2018
```

```python
vehicle_registration = vehicle_df_2018.drop(columns = ["Date", "Zip Code", "Duty"])
vehicle_registration
```

```python
vehicle_registration.to_csv("Resources/cleaned_vehicle_data_2018.csv", index=False)
```
